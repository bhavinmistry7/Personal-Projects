```python
import re
import json
import requests

from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://understat.com/player/8260'
```

```python
response = requests.get(url)
print(response.status_code)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
ugly_soup = str(soup)
```

```python
shots_data = re.search("var shotsData .*= JSON.parse\('(.*)'\)", ugly_soup).group(1)
```

```python
shots_data[:100]
```

```python
data = shots_data.encode('utf8').decode('unicode_escape')
```

```python
data[:100]
```

```python
data = json.loads(data)
```

```python
data
```

```python
df = pd.DataFrame(data)
haaland_2024_df = df[(df['player'] == 'Erling Haaland') & ((df['season'] == '2024'))]
```

```python
haaland_2024_df
```

```python
df.to_csv('haaland_2024_shots.csv', index=False)
```
