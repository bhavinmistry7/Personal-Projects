```python
import pandas as pd

import matplotlib.pyplot as plt
from mplsoccer import VerticalPitch

from fontTools.ttLib import TTCollection
```

```python
df = pd.read_csv('/Users/bhavin/Desktop/Soccer Data Analytics 2/Duran/duran_shots_so_far.csv')
```

```python
df.head()
```

```python
df['X'] = df['X'] * 100
df['Y'] = df['Y'] * 100
```

```python
total_shots = df.shape[0]
total_goals = df[df['result'] == 'Goal'].shape[0]
total_xG = df['xG'].sum()
xG_per_shot = total_xG / total_shots
points_average_distance = df['X'].mean()
actual_average_distance = 120 - (df['X'] * 1.2).mean()
```

```python
actual_average_distance
```

```python
background_color = '#0C0D0E'
```

```python
import matplotlib.font_manager as font_manager
font_path = '/Users/bhavin/Library/Fonts/Gotham Black Regular.ttf'
font_props=font_manager.FontProperties(fname=font_path)
```

```python
fig = plt.figure(figsize=(8,12))
fig.patch.set_facecolor(background_color)

ax1 = fig.add_axes([0, .7, 1, .2])
ax1.set_facecolor(background_color)
ax1.set_xlim(0,1)
ax1.set_ylim(0,1)


######################################Jhon Durán
ax1.text(
    x=.5,
    y=.85,
    s='Jhon Durán',
    fontsize=20,
    fontproperties=font_props,
    color='#fee505',
    ha='center'
)


######################################laliga shots text
ax1.text(
    x=.5,
    y=.68,
    s='All shots in the Premier League so far 2024-2025',
    fontsize=14,
    fontproperties=font_props,
    color='#a3c5e9',
    ha='center'
)

######################################low quality
ax1.text(
    x=.25,
    y=.5,
    s='Low Quality Chance',
    fontsize=12,
    fontproperties=font_props,
    color='#fee505',
    ha='center'
)


################################## all circles
#circle 1
ax1.scatter(
    x=.37,
    y=.53,
    s=100,
    color=background_color,
    edgecolor='white',
    linewidth=.8
)


#circle 2
ax1.scatter(
    x=.43,
    y=.53,
    s=200,
    color=background_color,
    edgecolor='white',
    linewidth=.8
)

#circle 3
ax1.scatter(
    x=.49,
    y=.53,
    s=300,
    color=background_color,
    edgecolor='white',
    linewidth=.8
)


#circle 4
ax1.scatter(
    x=.55,
    y=.53,
    s=400,
    color=background_color,
    edgecolor='white',
    linewidth=.8
)


#circle 5
ax1.scatter(
    x=.61,
    y=.53,
    s=500,
    color=background_color,
    edgecolor='white',
    linewidth=.8
)
###################################### end of circles

######################################high quality chances
ax1.text(
    x=.75,
    y=.5,
    s='High Quality Chance',
    fontsize=12,
    fontproperties=font_props,
    color='#fee505',
    ha='center'
)

##################################### goal or no goal
ax1.text(
    x=0.45, 
    y=0.28, 
    s=f'Goal', 
    fontsize=10, 
    fontproperties=font_props, 
    color='white', 
    ha='right'
)


ax1.scatter(
    x=0.47, 
    y=0.3, 
    s=100, 
    color='#fee505', 
    edgecolor='white', 
    linewidth=.8,
    alpha=.7
)


ax1.scatter(
    x=0.53, 
    y=0.3, 
    s=100, 
    color=background_color, 
    edgecolor='white', 
    linewidth=.8
)

ax1.text(
    x=0.55, 
    y=0.28, 
    s=f'No Goal', 
    fontsize=10, 
    fontproperties=font_props, 
    color='white', 
    ha='left'
)
##################################### goal or no goal ends


ax2 = fig.add_axes([.05, .25, .9, .5])
ax2.set_facecolor(background_color)

##################################### plotting the pitch
pitch = VerticalPitch(
    pitch_type='opta',
    half=True,
    pitch_color=background_color,
    pad_bottom=0.5,
    line_color='white',
    linewidth=.75,
    axis=True,
    label=True
)

pitch.draw(ax=ax2)
##################################### plotting the pitch ends

##################################### plotting average shot distance
ax2.scatter(x=90, y=points_average_distance,
            s=100, color='#a3c5e9', linewidth=.8)
ax2.plot([90,90], [100, points_average_distance,], color='white', linewidth=2)
ax2.text(
    x=90, y=points_average_distance -4, 
    s=f'Average Distance\n{actual_average_distance:.1f} yards',
    fontsize=10, fontproperties=font_props,
    color='#fee505', ha='center')


# Plot the furthest goal vs Everton
ax2.text(
    x=76, y=68, 
    s='Goal vs Everton',
    fontsize=10, fontproperties=font_props,
    color='#a3c5e9', ha='center')
ax2.text(
    x=76, y=66.5, 
    s=f'xG: 0.017',
    fontsize=10, fontproperties=font_props,
    color='#fee505', ha='center')

ax2.annotate(
    '',  # No text is needed for the arrow
    xy=(64, 72.5),  # The point on the plot you want the arrow to point to (X, Y of the goal)
    xytext=(73, 69.5),  # The starting point of the arrow (position near the text)
    arrowprops=dict(facecolor='white', edgecolor='white', arrowstyle='->', lw=1)  # Arrow properties
)

##################################### plotting the shots
for x in df.to_dict(orient='records'):
    pitch.scatter(
        x['X'],
        x['Y'],
        s=300* x['xG'],
        color='#fee505' if x['result'] == 'Goal' else background_color,
        ax=ax2,
        alpha=.7,
        linewidth=.8,
        edgecolor='white'
    )




ax3 = fig.add_axes([0, .2, 1, .05])
ax3.set_facecolor(background_color)

### shots
ax3.text(x=.25, y=.5, s='Shots', fontsize=20, fontproperties=font_props, color='#a3c5e9', ha='left')
ax3.text(x=.25, y=0, s=f'{total_shots}', fontsize=16, fontproperties=font_props, color='#fee505', ha='left')

### goals
ax3.text(x=.38, y=.5, s='Goals', fontsize=20, fontproperties=font_props, color='#a3c5e9', ha='left')
ax3.text(x=.38, y=0, s=f'{total_goals}', fontsize=16, fontproperties=font_props, color='#fee505', ha='left')

### xG
ax3.text(x=.53, y=.5, s='xG', fontsize=20, fontproperties=font_props, color='#a3c5e9', ha='left')
ax3.text(x=.53, y=0, s=f'{total_xG:.2f}', fontsize=16, fontproperties=font_props, color='#fee505', ha='left')

### xG per shot
ax3.text(x=.63, y=.5, s='xG/Shot', fontsize=20, fontproperties=font_props, color='#a3c5e9', ha='left')
ax3.text(x=.63, y=0, s=f'{xG_per_shot:.2f}', fontsize=16, fontproperties=font_props, color='#fee505', ha='left')



#Branding
ax3.text(x=0, y=.6, s='Bhavin', fontsize=25, fontproperties=font_props, color='white', ha='left')
ax3.text(x=0, y=.0, s='Mistry', fontsize=25, fontproperties=font_props, color='white', ha='left')

ax3.text(x=.85, y=-0.2, s='@bhavinho2', fontsize=12, fontproperties=font_props, color='white', ha='left')
```
