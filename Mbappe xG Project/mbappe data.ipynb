```python
import re
import json
import requests

from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://understat.com/player/3423'
```

```python
response = requests.get(url)
print(response.status_code)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
ugly_soup = str(soup)
```

```python
shots_data = re.search("var shotsData .*= JSON.parse\('(.*)'\)", ugly_soup).group(1)
```

```python
shots_data[:100]
```

```python
data = shots_data.encode('utf8').decode('unicode_escape')
```

```python
data[:100]
```

```python
data = json.loads(data)
```

```python
data
```

```python
df = pd.DataFrame(data)
madrid_df = df[(df['player'] == 'Kylian Mbappe-Lottin') & ((df['h_team'] == 'Real Madrid') | (df['a_team'] == 'Real Madrid'))]
```

```python
madrid_df
```

```python
madrid_df.to_csv('mbappe_real_madrid_data.csv', index=False)
```
